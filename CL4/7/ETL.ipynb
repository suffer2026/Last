```python
import pandas as pd

# Load dataset
df = pd.read_csv('Superstore.csv', encoding='ISO-8859-1')

# Preview data
print("Shape:", df.shape)
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
print(df.isnull().sum())
df = df.dropna(subset=['Sales', 'Profit'])

# Create new columns
df['Profit Margin'] = df['Profit'] / df['Sales']
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

# Standardize text columns
df['Customer Name'] = df['Customer Name'].str.title()
df['City'] = df['City'].str.title()
df['State'] = df['State'].str.title()

# Filter out invalid sales (if any)
df = df[df['Sales'] > 0]

df = df[['Order Date', 'Ship Date', 'Customer Name', 'Category', 'Sub-Category', 'Region', 'Sales', 'Profit']]

# Preview transformed data
print("Transformed Shape:", df.shape)
df.head()
```

```python
# Save cleaned dataset
output_path = 'Superstore_Cleaned.csv'
df.to_csv(output_path, index=False)

print("Saved to:", output_path)
```
